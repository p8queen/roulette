```python
# players at the same time
from random import randint as r
from random import shuffle as s

bola = lambda: r(0, 36)

def color(numero):
    rojo = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    if numero == 0:
        return 'verde'
    elif numero in rojo:
        return 'rojo'
    else:
        return 'negro'

def fibo(n):
    a, b = 1, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def lastNegatives(lista, cantidad):
    cantNegativos = 0
    for i in range(len(lista)-1,-1,-1):
        if lista[i]>0:
            return False
        else:
            cantNegativos += 1 
        if cantNegativos == cantidad:
            return True

bolas = [x for x in range(37)]*1000
s(bolas) #shuffle bolas
     
```

```python
# bet one dozen using fibonacci sequence   
j = 0
while j < 499:
    bet = 40
    win = False
    games1 = []
    i1 = 1

    for x in range(j,500):
        n = bolas[x]
        if 1 <= n <= 24:
            #print(n, '2nd dozen')
            win1 = True
        else:    
            #print(n, '1st dozen')
            win1 = False

        if win1:
            #print('You win', n)
            games1.append((bet//2)*fibo(i1))
            i1 = 1
            
        else:
            #print('You lose', n)
            games1.append(-(bet*fibo(i1)))
            i1 += 1
        if sum(games1) >= 120 or sum(games1) <= -130:
            break
    print(bolas[j:j+30])   
    print(' ----------- Totales ----------- ')
    print(f'Games1: {sum(games1)}')
    print('-------  Secuencia: --------')
    print(f'cantidad de juegos: {len(games1)}') 
    print(games1)
    print(' ----- FIN ----- ')
    print()

    j = x+1

```

```python
# triple martingala 
# bet two dozen using fibonacci sequence   
bolas = [x for x in range(37)]*1000
s(bolas) #shuffle bolas

j = 0
while j < 499:
    bet = 2
    win = False
    games1 = []
    factor = 1

    for x in range(j,500):
        n = bolas[x]
        if 1 <= n <= 24:
            #print(n, '2nd dozen')
            win1 = True
        else:    
            #print(n, '1st dozen')
            win1 = False

        if win1:
            #print('You win', n)
            games1.append((factor*bet//2))
            factor = 1
            
        else:
            #print('You lose', n)
            games1.append(-(factor*bet))
            factor *= 3
        if sum(games1) >= 80 or sum(games1) <= -231:
            break
    print(bolas[j:j+30])   
    print(' ----------- Totales ----------- ')
    print(f'Games1: {sum(games1)}')
    print('-------  Secuencia: --------')
    print(f'cantidad de juegos: {len(games1)}') 
    print(games1)
    print(' ----- FIN ----- ')
    print()

    j = x+1
```
