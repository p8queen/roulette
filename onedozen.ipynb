Bet ine dozen, duplicate bet if win. 

```python
from random import shuffle as s


def color(numero):
    rojo = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    if numero == 0:
        return 'verde'
    elif numero in rojo:
        return 'rojo'
    else:
        return 'negro'

#MAIN

bolas = [x for x in range(37)]*1000
s(bolas) #shuffle bolas

# bet one dozen using fibonacci sequence   
j = 0
while j < 499:
    bet = 20
    win = False
    games1 = []
    i1 = 1
    bet_init = 20
    ganados  = 0

    for x in range(j,500):
        n = bolas[x]
        if 13 <= n <= 24:
            #print(n, '2nd dozen')
            win1 = True
        else:    
            #print(n, '1st dozen')
            win1 = False

        if win1:
            #print('You win', n)
            games1.append(bet*2)
            ganados = 1 if ganados == 0 else 0
            bet = bet_init * 2 if ganados == 1 else bet_init
        else:
            #print('You lose', n)
            games1.append(-(bet))
            bet = bet_init
            ganados = 0 

            
        if sum(games1) >= 120 or sum(games1) <= -130:
            break
    print(f'bolas: {bolas[j:j+30]}')   
    print(' ----------- Totales ----------- ')
    print(f'Games1: {sum(games1)}')
    print('-------  Secuencia: --------')
    print(f'cantidad de juegos: {len(games1)}') 
    print(games1)
    print(' ----- FIN ----- ')
    print()

    j = x+1

```
