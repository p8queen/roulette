```python
# players at the same time
from random import randint as r
from random import shuffle as s

bola = lambda: r(0, 36)

def color(numero):
    rojo = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    if numero == 0:
        return 'verde'
    elif numero in rojo:
        return 'rojo'
    else:
        return 'negro'

def fibo(n):
    a, b = 1, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def lastNegatives(lista, cantidad):
    cantNegativos = 0
    for i in range(len(lista)-1,-1,-1):
        if lista[i]>0:
            return False
        else:
            cantNegativos += 1 
        if cantNegativos == cantidad:
            return True
```

```python
# bet one dozen using fibonacci sequence   
bet = 5
win = False
games1 = []
games2 = []
progre = []
i1 = 1
i2 = 1
p = 1 # system 1-2-3
bolas = [x for x in range(37)]*1000
s(bolas) #shuffle bolas
for x in range(90):
    n = bolas[x]
    if 1 <= n < 13:
        #print(n, '2nd dozen')
        win1 = True
    else:    
        #print(n, '1st dozen')
        win1 = False
    if win1:
        #print('You win', n)
        games1.append(2*bet*fibo(i1))
        progre.append(2*bet*p)
        i1 = 1
        p -= 1
        p = max(1, p)
    else:
        #print('You lose', n)
        games1.append(-(bet*fibo(i1)))
        progre.append(-(bet*p))
        i1 += 1
        p += 1
    
    if 13 <= n < 25:
        #print(n, '2nd dozen')
        win2 = True
    else:    
        #print(n, '1st dozen')
        win2 = False
    if win2:
        #print('You win', n)
        games2.append(2*bet*fibo(i2))
        progre.append(2*bet*p)
        i2 = 1
        p -= 1
        p = max(1, p)
    else:
        #print('You lose', n)
        games2.append(-(bet*fibo(i2)))
        progre.append(-(bet*p))
        i2 += 1
        p += 1
    #print(f'Games: {games}, Total: {sum(games)}')
    #print(f'Progress: {progre}, Total: {sum(progre)}')
    if sum(games1) + sum(games2) >= 120:
        break
    if lastNegatives(games1,15) or lastNegatives(games2,15):
        break


# max drowdown 
def drowdown(lista):
    j = 1
    while j < len(lista):
        print(sum(lista[:j]), end=', ')
        j += 1

print(' ----------- Totales ----------- ')
print(f'Games1: {sum(games1)}')
print(f'Games2: {sum(games2)}')
print()

print('-------  Secuencia: --------')
print(games1)
print()
print(games2)
print()

print('-------  drowdown: --------')
drowdown(games1)
print()
drowdown(games2)
```
