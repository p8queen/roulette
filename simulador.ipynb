```python
# players at the same time
from random import randint as r
from random import shuffle as s
import regex as re

bola = lambda: r(0, 36)

def fibo(n):
    a, b = 1, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def color(numero):
    rojo = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    if numero == 0:
        return 'verde'
    elif numero in rojo:
        return 'rojo'
    else:
        return 'negro'

def fibo(n):
    a, b = 1, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def lastNegatives(lista, cantidad):
    cantNegativos = 0
    for i in range(len(lista)-1,-1,-1):
        if lista[i]>0:
            return False
        else:
            cantNegativos += 1 
        if cantNegativos == cantidad:
            return True

def columnas(numero):
    if numero == 0:
        return 0
    columna = numero%3 
    columna = 3 if columna == 0 else columna
    return columna

def bet(bola, posicion, importe):
    #bola es rnd, posicion tablero (numero, color, coloumna), importe
    # usar regex para match numero 

    pattern_number = r'^(0|[1-9]|[1-2][0-9]|3[0-6])$'
    pattern_columna = r'^columna-\d+'
    pattern_color = r'^(rojo|negro)$'

    
    if re.match(pattern_number, str(posicion)):
        print(f"{posicion} is a number match")
        if posicion == bola:
            print(f'Cilindro: {bola}, {posicion} Win Pleno')
            return 1
        else:
            print(f'Cilindro: {bola}, {posicion} Loosse Pleno')
            return 0
    elif re.match(pattern_color, posicion):
        if color(bola) == posicion:
            print(f'Cilindro: {bola}, Win Color: {posicion}')
            return 1
        else:
            print(f'Cilindro: {bola}, Lose Color: {posicion}')
            return 0
    
    elif re.match(pattern_columna, posicion):
        numeroColumna = int(posicion.split('-')[1])
        if columnas(bola) == numeroColumna:
            print(f'Cilindro: {bola}, Win Columna: {numeroColumna}')
            return 1
        else:
            print(f'Cilindro: {bola}, Lose Columna: {numeroColumna}')
            return 0
    else:
        print('No match')
        return -1




```

```python
from random import randint as r
from random import shuffle as s

def getColor(numero):
    rojo = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    if numero == 0:
        return 'verde'
    elif numero in rojo:
        return 'rojo'
    else:
        return 'negro'
    
def getDocena(numero):
    if numero == 0:
        return -1
    elif numero < 13:
        return 1
    elif numero < 25:
        return 2
    else:
        return 3

def getColumnas(numero):
    if numero == 0:
        return 0
    columna = numero%3 
    columna = 3 if columna == 0 else columna
    return columna
   
def getBola(bolas, posicion):
    numero = bolas[posicion]
    color = getColor(numero)
    docena = getDocena(numero)
    columna = getColumnas(numero)
    resultado = {"numero": numero, "color": color, "docena": docena, "columna": columna}
    return resultado

bolas = [x for x in range(37)]*1000
s(bolas) #shuffle bolas

def betColor(color, objBola):
    if objBola['color'] == color: 
        return 1
    return 0

balance = []
sumatoriaBalance = []
i = 1
for x in range(20):
    betFibo = fibo(i)
    bola = getBola(bolas, x)
    resultado = betColor('rojo', bola)
    if resultado == 1: #win
        i = max(1, i-1)
        balance.append(betFibo)
    else:
        i += 1
        balance.append(-betFibo)
    sumatoriaBalance.append(sum(balance))
    #print(f'Bola: {bola}')

print(f'Balance: {balance}')
print(f'Sumatoria Balance: {sumatoriaBalance}')
print(f'Sumatoria Total Balance: {sum(balance)}')
```
